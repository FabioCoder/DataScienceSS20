## Exercise 1
Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: 

* X label name = Month Number
* Y label name = Total profit


```python
data.plot(x='month_number',y='total_profit')
plt.xlabel('Month Number')
plt.ylabel('Total profit')
 
```

## Exercise 2
Get Total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

* Line Style dotted and Line color should be red (https://matplotlib.org/2.1.1/api/_as_gen/matplotlib.pyplot.plot.html)
* Show legend at the lower right location. (https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html#matplotlib.pyplot.legend)
* X label name = Month Number
* Y label name = Sold units number
* Add a circle marker. (https://stackoverflow.com/questions/8409095/matplotlib-set-markers-for-individual-points-on-a-line)
* Line marker color as read
* Line width should be 3


```python
#Parameter style, linewidth und color bestimmen das Aussehen der Linie
#Der Parameter marker setzt den Line Marker 
data.plot(x='month_number',y='total_units', style=':', color='red', linewidth='3', marker='o')
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
#Location der Legende
plt.legend(loc='lower right')
```

## Exercise 3
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).

Auswahl von bestimmten Columns: https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html

```python
for column in data.loc[:,'facecream':'moisturizer']:
    plt.plot(data.month_number,data[column], label=column)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

## Exercise 6
Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

***Note:*** use ``data[item].sum()`` to compute sums over columns 

```python
data['toothpaste'].sum()
```

```python
unitsPerProduct = data.loc[:,'facecream':'moisturizer'].sum()
total_units = unitsPerProduct.sum()
df = unitsPerProduct / total_units

df.plot.pie(autopct='%1.1f%%')
```
