```python
#3.5: compute the matrix inverse of A

#AInv ist die Inverse Matrix von A
AInv = np.linalg.inv(A)

#Beweis
E = A.dot(AInv)
print(np.diag(E))

#Aufgrund von rundungsungenauigkeiten ist alles rechts und links von der diagonale nur nahezu 0 
mask = E > 0.0000000001
print(mask)
```
